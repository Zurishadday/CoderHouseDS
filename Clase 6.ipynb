* **Time stamp o marca de tiempo**: representan un punto en el tiempo. Por ejemplo, fecha y hora.
* **Per铆odo**: representan un intervalo de tiempo. Por ejemplo, frecuencia anual, mensual, semanal (semana del a帽o)
* **Duraci贸n**: representa una duraci贸n medida en tiempo, es el diferencial de un tiempo inicial a uno final. En pocas palabras el diferencial.
